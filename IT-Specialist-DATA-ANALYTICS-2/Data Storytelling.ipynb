```python
df = pd.read_csv('Titanic.csv')
```

```python
df.describe()
```

```python
# Create a DataFrame showing the count of survived and not survived passengers by sex
survival_by_sex = df.groupby(["Survived", "Sex"]).size().unstack()

# Print the results
print("Number of Survived Passengers:")
print(survival_by_sex)
```

This data shows the number of passengers who survived the Titanic disaster, sorted by gender. Overall, there were 109 male passengers who survived and 233 female passengers who survived. Looking at this data, it can be seen that there is a significant difference in survival rates between men and women. The number of female passengers who survived was far greater than male passengers.
